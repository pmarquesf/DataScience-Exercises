#### b-) Ler o dataset "*Hourly_wages/hourly_wages.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e 5 últimas linhas.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
wage = pd.read_csv('DSBD/Datasets/Hourly_wages/hourly_wages.csv')
wage
```

#### c-) Mostrar os histogramas dos atributos: *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
plt.subplot(2,3,1)
plt.hist(wage['education_yrs'],color='red')
plt.title('education_yrs:')

plt.subplot(2,3,2)
plt.hist(wage['experience_yrs'], color='blue')
plt.title('experience_yrs:')

plt.subplot(2,3,3)
plt.hist(wage['age'], color='yellow')
plt.title('Age:')

plt.show
```

#### d-) Mostrar os gráficos de setores (*pie chart*) de cada um dos atributos: *union*, *female*, *marr*, *south*, *manufacturing* e *construction* num mesmo quadro de plotagem (usando subplots).

```
from pandas.core.internals import construction
union = ['filiado', 'Não filiado'] 
data = [17.98, 82.02] 
plt.subplot(2,3,1)
plt.pie(data, labels = union,colors=['blue','darkblue'])
plt.title('Situação:')

female = ['Homem', 'Mulher'] 
data = [54.11, 45.89]
plt.subplot(2,3,2)
plt.pie(data, labels = female, colors=['orange','red'])
plt.title('Gênero:')

marr = ['Casado', 'Solteiro'] 
data = [65.54, 34.46] 
plt.subplot(2,3,3)
plt.pie(data, labels = marr, colors=['blue','lightpink'])
plt.title('Estado civíl:')

south = ['Norte', 'Sul'] 
data = [70.78, 29.22] 
plt.subplot(2,3,4)
plt.pie(data, labels = south,colors=['green','lightgreen'])
plt.title('Região:')

manufacturing = ['Maquinof.', 'Manuf.'] 
data = [88.46, 18.54] 
plt.subplot(2,3,5)
plt.pie(data, labels = manufacturing, colors=['orange','black'])
plt.title('Tipo de Trabalho:')

construction = ['Parado', 'Contínuo'] 
data = [95.5, 4.5] 
plt.subplot(2,3,6)
plt.pie(data, labels = construction, colors=['grey','darkgrey'])
plt.title('Construção:')
plt.show()
```

#### e-) Mostrar os boxplots das variáveis *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.boxplot(ax=axes[0], data=wage, color='red', x='education_yrs')
sns.boxplot(ax=axes[1], data=wage, color='blue', x='experience_yrs')
sns.boxplot(ax=axes[2], data=wage, color='orange', x='age')
```
