#Exercício 4.2 - ESTIMAÇÃO INTERVALAR

De uma amostra aleatória de 50 dias deste ano, os preços dos combustíveis nos EUA tiveram uma média de US 4,28. Suponha que o desvio padrão  populacional é de US 0,21. Determine o intervalo de confiança (estimativa intervalar) para a média populacional do preço dos combustíveis:

a)Com 95% de confiança;
b)Com 99% de confiança;

---
---


Importação de Bibliotecas

```python
import numpy as np
import seaborn as sns
import math
from scipy import stats
from scipy.stats import norm, zscore
```

A) com 95% de confiança

```python
n = 50
media = 4.28
numeroconfianca = 0.95
desvio_padrao = 0.21
intervalo = norm.interval(numeroconfianca, media, desvio_padrao/math.sqrt(n))
intervalo
```

B) Com 99% de confiança

```python
n = 50
media = 4.28
numeroconfianca = 0.99
desvio_padrao = 0.21
intervalo = norm.interval(numeroconfianca, media, desvio_padrao/math.sqrt(n))
intervalo
```
