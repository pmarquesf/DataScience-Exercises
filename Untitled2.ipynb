````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
````

##leitura do dataset

````python
df = pd.read_csv('/content/DSBD/Datasets/Empregabilidade/empregabilidade.csv')
df
````

###verificação de dados ausentes

````python
df.isnull().sum()
````

###elimina dados ausentes

````python
column_mean = df['ssc_p'].mean()
df['ssc_p'] = df['ssc_p'].fillna(value = column_mean)
df['ssc_p']
````

````python
column_mean = df['mba_p'].mean()
df['mba_p'] = df['mba_p'].fillna(value = column_mean)
df['mba_p']
````

````python
column_mean = df['salary'].mean()
df['salary'] = df['salary'].fillna(value = column_mean)
df['salary']
````

##Criação variável X parafatores que podem influenciar na contratação

````python
X = df.iloc[:, 1:13] 
X
````

criação de variável y para o target

````python
y = df.iloc[:, -2]
y
````

##codificando as colunas char do X

````python
labelencoder = LabelEncoder()
X['gender'] = labelencoder.fit_transform(X['gender'])
X['ssc_b'] = labelencoder.fit_transform(X['ssc_b'])
X['hsc_b'] = labelencoder.fit_transform(X['hsc_b'])
X['hsc_s'] = labelencoder.fit_transform(X['hsc_s'])
X['degree_t'] = labelencoder.fit_transform(X['degree_t'])
X['workex'] = labelencoder.fit_transform(X['workex'])
X['specialisation'] = labelencoder.fit_transform(X['specialisation'])
X

````

##codificando target Y

````python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(y)
y
````

##Dividindo em train & test

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

#Regressão Logística

````python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver = 'liblinear')
logreg.fit(X_train,y_train)
classificacoes_logreg = logreg.predict(X_test)
acertos_LogReg = metrics.accuracy_score(y_test,classificacoes_logreg)
acertos_LogReg = metrics.accuracy_score(y_test,classificacoes_logreg)
print('Taxa de acerto = ',round(acertos_LogReg*100,2),"%\n")
print('Matriz de confusão:\n')
cm_logreg = ConfusionMatrixDisplay.from_estimator(logreg, X_test, y_test)
plt.show()
````

#KNN - K-Nearest Neighbors 


````python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=20) 
knn.fit(X_train,y_train)
classificacoes_knn = knn.predict(X_test)
acertos_knn = metrics.accuracy_score(y_test,classificacoes_knn)
print('Taxa de acerto = ',round(acertos_knn*100,2),"%\n")
print('Matriz de confusão:\n')
cm_knn = ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test)
plt.show()
````

#SVM

````python
from sklearn import svm
svm = svm.SVC()
svm.fit(X_train, y_train)
classificacoes_svm = svm.predict(X_test)
acertos_svm = metrics.accuracy_score(y_test, classificacoes_svm)
print('Taxa de acerto = ',round(acertos_svm*100,2),"%\n")
print('Matriz de confusão:\n')
cm_svm = ConfusionMatrixDisplay.from_estimator(svm, X_test, y_test)
plt.show()
````

#Random Forest

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 10, random_state = 42)
rf.fit(X_train, y_train)
classificacoes_rf = rf.predict(X_test)
acertos_rf = metrics.accuracy_score(y_test, classificacoes_rf)
print('Taxa de acerto = ',round(acertos_rf*100,2),"%\n")
print('Matriz de confusão:\n')
cm_rf = ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)
plt.show()
````

#Rede Neural de Perceptrons Multicamadas (deve ser implementada utilizando o Keras/Tensorflow)


##Criação do modelo de rede Neural

````python
from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import cross_val_score

model = Sequential()
model.add(Dense(24, input_dim=12, kernel_initializer = 'he_uniform', activation='relu'))
model.add(Dense(12, kernel_initializer = 'he_uniform', activation='relu'))
model.add(Dense(6, kernel_initializer = 'he_uniform', activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
````

##Compila o Modelo

````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

##Treina o modelo

````python
history = model.fit(X_train, y_train, epochs=100, batch_size = 8, validation_data=(X_test, y_test))
````

Gráfico da função LOSS

````python
#função Loss
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'], '')
plt.xlabel("time")
plt.ylabel('loss')
plt.title('Funçaõ loss')
plt.legend(['loss', 'val_loss'])
plt.show()
````

##Gráfico Accuracy

````python
#accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'], '')
plt.xlabel("Épocas")
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.legend(['accuracy', 'val_accuracy'])
plt.show()
````

##Classificação dos dados

````python
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
````
