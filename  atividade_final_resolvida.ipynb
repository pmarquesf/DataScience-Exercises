````python
#4 Análise gráfica

#a)sobrevivencia por sexo
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sns.boxplot(x = 'Survived', y = 'Sex', data=df)
plt.title('Sobrevivência entre sexos')
````

De acordo Com o gráfico, mulheres possuíram muito mais chance de sobreviver

````python
#b)A idade influenciou na sobrevivência?

idade = sns.FacetGrid(df, col='Survived')
idade.map(sns.distplot, 'Age')
````

A partir dos gráfico se percebe um pico do lado para sobreviventes na parte de ciranças, porvando que crianças tem mais chances de sobreviver 

````python
#c)O local de embarque tem relação com o preço da passagem?
fig, ax = plt.subplots()
sns.barplot(x = 'Embarked', y = 'Fare', data=df)
plt.title('Relação Embarque e valor da Passsagem')
````

Como se pode perceber, passageiros que embarcaram em "C" pagaram uma tarifa mais cara
